```python
import pandas as pd
import numpy as np
```

```python
player_2016_2017 = pd.read_excel('NBA-2016-2017-Player-BoxScore-Dataset.xlsx')

player_2016_2017 = player_2016_2017.rename(index=str, columns={'DATA SET':'DATASET','PLAYER FULL NAME':'PLAYER','OWN TEAM':'TEAM','OPP TEAM':'OPPONENT','VENUE (R/H)':'ROAD_HOME','MIN':'MINUTES'})

player_2016_2017['DT_Double'] = pd.Series(np.zeros(player_2016_2017.shape[0]))
player_2016_2017.DT_Double.fillna(value=0.0,inplace=True)
```

```python
#This counts the number of categories the player had more than 10
#This takes a long time, consider rewriting
categories = ['PTS', 'TOT', 'A', 'BL', 'ST']
for cat in categories:
    for i,x in enumerate(player_2016_2017[cat]):
        if x >=10:
            player_2016_2017['DT_Double'][i] += 1
```

```python
#This adjusts the 'DT_Double column to the appropriate multiplier
#This takes a long time, consider rewriting
for i,x in enumerate(player_2016_2017['DT_Double']):
    if x == 1 or x == 2:
        player_2016_2017['DT_Double'][i] -= 1
```

```python
player_2016_2017['DraftKings_PTS'] = player_2016_2017['PTS'] + (player_2016_2017['3P'] * 0.5) + (player_2016_2017['TOT'] * 1.25) + (player_2016_2017['A'] * 1.5) + (player_2016_2017['ST'] * 2) + (player_2016_2017['BL'] * 2) - (player_2016_2017['TO'] * 0.5) + (player_2016_2017['DT_Double'] * 1.5) 
```
