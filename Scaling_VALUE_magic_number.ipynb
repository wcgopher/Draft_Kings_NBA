```python
import pandas as pd
import numpy as np
%matplotlib inline

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import scale
```

```python
data = pd.read_pickle('player_2016_2018.pkl')
data = data.dropna()
```

```python
data['DK_SAL'] = pd.to_numeric(data['DK_SAL'])
data['VALUE'] = pd.to_numeric(data['VALUE'])
data['FAVORITE'] = data['FAVORITE'].astype(int)
data['SPREAD_10+'] = data['SPREAD_10+'].astype(int)
data['magic_number'] = data['DraftKings_PTS'] * data['VALUE']
# data['Log_VALUE'] = np.log10(data['VALUE'])
# data['Log_magic_number'] = np.log10(data['magic_number'])
```

```python
standard_scale = StandardScaler()
MinMax_scale = MinMaxScaler()
```

```python
data['VALUE'] = scale(data['VALUE'])
data['magic_number'] = scale(data['magic_number'])
```

```python
data.sort_values(by = 'magic_number', ascending=False)
```

```python
data.VALUE.plot(kind = 'hist', figsize=(13,7),logx=False)
```
