```python
import pandas as pd
import numpy as np
```

```python
player_2017_2018 = pd.read_pickle('player_2017_2018_w_hand_calc_DKPTS.pkl')
team_2017_2018 = pd.read_pickle('team_2017_18_w_opponent_stats.pkl')
player_2017_2018.columns
team_2017_2018.columns
```

```python
player_2017_2018.head()
```

```python
df = player_2017_2018[['DATE','TEAM','POSITION','DraftKings_PTS']]
```

```python
df.head()
```

```python
df.POSITION.unique()
```

```python
positional_pts = df.groupby(['DATE','TEAM','POSITION']).sum()
```

```python
positional_pts = positional_pts.unstack(level=-1)
```

```python
positional_pts.head()
```

```python
positional_pts = positional_pts.xs('DraftKings_PTS', axis=1, drop_level=True)
```

```python
positional_pts = positional_pts.reset_index(level=1)
```

```python
positional_pts = positional_pts.reset_index()
```

```python
positional_pts = positional_pts.rename(columns={'TEAM':'TEAMS','C':'C_DK_PTS','F':'F_DK_PTS','G':'G_DK_PTS','PF':'PF_DK_PTS','PG':'PG_DK_PTS','SF':'SF_DK_PTS','SG':'SG_DK_PTS'})
```

```python
positional_pts = positional_pts.set_index(['DATE','TEAMS'])
```

```python
positional_pts.head()
```

```python
result = team_2017_2018.join(positional_pts,on=['DATE','TEAMS'])
```

```python
result.head(30)
```

```python
result.to_pickle('team_2017_18_w_opponent_stats_and_DvP.pkl')
```
