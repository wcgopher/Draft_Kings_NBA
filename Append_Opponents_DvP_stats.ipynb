```python
import pandas as pd
import numpy as np
```

```python
team_2017_2018 = pd.read_pickle('team_2017_18_w_opponent_stats_and_DvP.pkl')
```

```python
team_2017_2018.columns
```

```python
#Fill NaN spots with 0 under the DK_PTS columns
team_2017_2018[['C_DK_PTS',
       'F_DK_PTS', 'G_DK_PTS', 'PF_DK_PTS', 'PG_DK_PTS', 'SF_DK_PTS',
       'SG_DK_PTS']] = team_2017_2018[['C_DK_PTS',
       'F_DK_PTS', 'G_DK_PTS', 'PF_DK_PTS', 'PG_DK_PTS', 'SF_DK_PTS',
       'SG_DK_PTS']].fillna(value=0)
```

```python
#Combine G and SG columns and Fand PF columns. Drop F and G columns
team_2017_2018['PF_DK_PTS'] = team_2017_2018['PF_DK_PTS'] + team_2017_2018['F_DK_PTS']
team_2017_2018['SG_DK_PTS'] = team_2017_2018['SG_DK_PTS'] + team_2017_2018['G_DK_PTS']
team_2017_2018 = team_2017_2018.drop(['F_DK_PTS','G_DK_PTS'],axis=1)
```

```python
team_2017_2018.tail(20)
```

```python
#create list of the statistics you want to get the opponent's value for
stats = ['C_DK_PTS', 'PF_DK_PTS', 'PG_DK_PTS', 'SF_DK_PTS','SG_DK_PTS']
len(stats)
```

```python
#function that appends a DF of the stat to a list
def add_opponent_stat_column(stat,frames):
    stat_df = pd.DataFrame(team_2017_2018[[stat]])
    stat_list = []
    for i,row in stat_df.iterrows():
        stat_list.append(row[0])
    opponent_stat_list = []
    
    for i,team in enumerate(stat_list):
        if i % 2 == 0:
            opponent_stat_list.append(stat_list[i+1])
        else:
            opponent_stat_list.append(stat_list[i-1])

    opponent_stat_list_df = pd.DataFrame(opponent_stat_list,columns=['opp_'+stat])

    frames.append(opponent_stat_list_df)


    return frames
```

```python
#initialize frames list with origional DF
frames = [team_2017_2018]
#Loop through that stats you want the opponents value of and that stat DF will be appended to frames list
for stat in stats:
    frames = add_opponent_stat_column(stat,frames)
len(frames)
```

```python
#Concat all frames in frames list
team_2017_2018 = pd.concat(frames,axis=1)
```

```python
team_2017_2018 = team_2017_2018.rename(columns={'opp_C_DK_PTS':'opp_C_allowed',
       'opp_PF_DK_PTS':'opp_PF_allowed', 'opp_PG_DK_PTS':'opp_PG_allowed',
                               'opp_SF_DK_PTS':'opp_SF_allowed','opp_SG_DK_PTS':'opp_SG_allowed'})
team_2017_2018.columns
```

```python
team_2017_2018.head()
```

```python
#create small DFs with pace, oeff, and deff
PG_DK_PTS_allowed = team_2017_2018[['TEAMS','opp_PG_allowed']]
SG_DK_PTS_allowed = team_2017_2018[['TEAMS','opp_SG_allowed']]
SF_DK_PTS_allowed = team_2017_2018[['TEAMS','opp_SF_allowed']]
PF_DK_PTS_allowed = team_2017_2018[['TEAMS','opp_PF_allowed']]
C_DK_PTS_allowed = team_2017_2018[['TEAMS','opp_C_allowed']]

#Create list with all team names
teams = team_2017_2018.TEAMS.unique()
```

```python
def join_rolling_opp_PG_allowed(teams,stat_df):
    result = []
    for team in teams:
        df = pd.DataFrame(PG_DK_PTS_allowed.loc[lambda df: stat_df.TEAMS == team])
        df['opp_PG_allowed_roll3'] = df['opp_PG_allowed'].rolling(3, min_periods=1).mean().shift()
        df['opp_PG_allowed_roll5'] = df['opp_PG_allowed'].rolling(5, min_periods=1).mean().shift()
        df['opp_PG_allowed_roll10'] = df['opp_PG_allowed'].rolling(10, min_periods=1).mean().shift()
        df['opp_PG_allowed_roll25'] = df['opp_PG_allowed'].rolling(25, min_periods=1).mean().shift()
        result.append(df)
    return result

def join_rolling_opp_SG_allowed(teams,stat_df):
    result = []
    for team in teams:
        df = pd.DataFrame(SG_DK_PTS_allowed.loc[lambda df: stat_df.TEAMS == team])
        df['opp_SG_allowed_roll3'] = df['opp_SG_allowed'].rolling(3, min_periods=1).mean().shift()
        df['opp_SG_allowed_roll5'] = df['opp_SG_allowed'].rolling(5, min_periods=1).mean().shift()
        df['opp_SG_allowed_roll10'] = df['opp_SG_allowed'].rolling(10, min_periods=1).mean().shift()
        df['opp_SG_allowed_roll25'] = df['opp_SG_allowed'].rolling(25, min_periods=1).mean().shift()
        result.append(df)
    return result

def join_rolling_opp_SF_allowed(teams,stat_df):
    result = []
    for team in teams:
        df = pd.DataFrame(SF_DK_PTS_allowed.loc[lambda df: stat_df.TEAMS == team])
        df['opp_SF_allowed_roll3'] = df['opp_SF_allowed'].rolling(3, min_periods=1).mean().shift()
        df['opp_SF_allowed_roll5'] = df['opp_SF_allowed'].rolling(5, min_periods=1).mean().shift()
        df['opp_SF_allowed_roll10'] = df['opp_SF_allowed'].rolling(10, min_periods=1).mean().shift()
        df['opp_SF_allowed_roll25'] = df['opp_SF_allowed'].rolling(25, min_periods=1).mean().shift()
        result.append(df)
    return result

def join_rolling_opp_PF_allowed(teams,stat_df):
    result = []
    for team in teams:
        df = pd.DataFrame(PF_DK_PTS_allowed.loc[lambda df: stat_df.TEAMS == team])
        df['opp_PF_allowed_roll3'] = df['opp_PF_allowed'].rolling(3, min_periods=1).mean().shift()
        df['opp_PF_allowed_roll5'] = df['opp_PF_allowed'].rolling(5, min_periods=1).mean().shift()
        df['opp_PF_allowed_roll10'] = df['opp_PF_allowed'].rolling(10, min_periods=1).mean().shift()
        df['opp_PF_allowed_roll25'] = df['opp_PF_allowed'].rolling(25, min_periods=1).mean().shift()
        result.append(df)
    return result

def join_rolling_opp_C_allowed(teams,stat_df):
    result = []
    for team in teams:
        df = pd.DataFrame(C_DK_PTS_allowed.loc[lambda df: stat_df.TEAMS == team])
        df['opp_C_allowed_roll3'] = df['opp_C_allowed'].rolling(3, min_periods=1).mean().shift()
        df['opp_C_allowed_roll5'] = df['opp_C_allowed'].rolling(5, min_periods=1).mean().shift()
        df['opp_C_allowed_roll10'] = df['opp_C_allowed'].rolling(10, min_periods=1).mean().shift()
        df['opp_C_allowed_roll25'] = df['opp_C_allowed'].rolling(25, min_periods=1).mean().shift()
        result.append(df)
    return result
```

```python
#execute the functions and save assign them variables
rolling_opp_PG_allowed = join_rolling_opp_PG_allowed(teams,PG_DK_PTS_allowed)
rolling_opp_SG_allowed = join_rolling_opp_SG_allowed(teams,SG_DK_PTS_allowed)
rolling_opp_SF_allowed = join_rolling_opp_SF_allowed(teams,SF_DK_PTS_allowed)
rolling_opp_PF_allowed = join_rolling_opp_PF_allowed(teams,PF_DK_PTS_allowed)
rolling_opp_C_allowed = join_rolling_opp_C_allowed(teams,C_DK_PTS_allowed)
```

```python
#Concat each list of DFs
rolling_opp_PG_allowed = pd.concat(rolling_opp_PG_allowed)
rolling_opp_SG_allowed = pd.concat(rolling_opp_SG_allowed)
rolling_opp_SF_allowed = pd.concat(rolling_opp_SF_allowed)
rolling_opp_PF_allowed = pd.concat(rolling_opp_PF_allowed)
rolling_opp_C_allowed = pd.concat(rolling_opp_C_allowed)
#Reorder the rows by original index
rolling_opp_PG_allowed = rolling_opp_PG_allowed.sort_index()
rolling_opp_SG_allowed = rolling_opp_SG_allowed.sort_index()
rolling_opp_SF_allowed = rolling_opp_SF_allowed.sort_index()
rolling_opp_PF_allowed = rolling_opp_PF_allowed.sort_index()
rolling_opp_C_allowed = rolling_opp_C_allowed.sort_index()
#Drop the 'TEAMS' column and the stat column
rolling_opp_PG_allowed = rolling_opp_PG_allowed.drop(['TEAMS','opp_PG_allowed'],axis=1)
rolling_opp_SG_allowed = rolling_opp_SG_allowed.drop(['TEAMS','opp_SG_allowed'],axis=1)
rolling_opp_SF_allowed = rolling_opp_SF_allowed.drop(['TEAMS','opp_SF_allowed'],axis=1)
rolling_opp_PF_allowed = rolling_opp_PF_allowed.drop(['TEAMS','opp_PF_allowed'],axis=1)
rolling_opp_C_allowed = rolling_opp_C_allowed.drop(['TEAMS','opp_C_allowed'],axis=1)
```

```python
#join the new DFs to the origional team stat DF
team_2017_2018 = team_2017_2018.join(rolling_opp_PG_allowed)
team_2017_2018 = team_2017_2018.join(rolling_opp_SG_allowed)
team_2017_2018 = team_2017_2018.join(rolling_opp_SF_allowed)
team_2017_2018 = team_2017_2018.join(rolling_opp_PF_allowed)
team_2017_2018 = team_2017_2018.join(rolling_opp_C_allowed)
```
