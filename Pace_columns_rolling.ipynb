```python
import pandas as pd
import numpy as np
%matplotlib inline

team_2017_2018 = pd.read_excel('/Users/willhall/Desktop/python_projects/DraftKings/2017-2018_NBA_Box_Score_Team-Stats.xlsx')
team_2017_2018 = team_2017_2018.drop(['FG', 'FGA', '3P',
       '3PA', 'FT', 'FTA', 'OR', 'DR', 'TOT', 'A', 'PF', 'ST', 'TO', 'TO TO',
       'BL','1Q', '2Q', '3Q', '4Q', 'OT1',
       'OT2', 'OT3', 'OT4','STARTING LINEUPS', 'Unnamed: 36',
       'Unnamed: 37', 'Unnamed: 38', 'Unnamed: 39', 'MAIN REF', 'CREW','OPENING ODDS', 'OPENING SPREAD',
       'OPENING TOTAL', 'MOVEMENTS', 'CLOSING ODDS', 'MONEYLINE', 'HALFTIME',
       'BOX SCORE', 'POSS','ODDS'], axis=1)

teams = team_2017_2018.TEAMS.unique()

for team in teams:
    team_2017_2018 = join_pace(team,team_2017_2018)


team_2017_2018['pace_roll2_combo'] = team_2017_2018['pace_roll2'] + team_2017_2018['pace_roll2Cleveland'] + team_2017_2018['pace_roll2Houston'] + team_2017_2018['pace_roll2Golden State'] + team_2017_2018['pace_roll2Charlotte'] + team_2017_2018['pace_roll2Detroit'] + team_2017_2018['pace_roll2Brooklyn'] + team_2017_2018['pace_roll2Indiana'] + team_2017_2018['pace_roll2Miami'] + team_2017_2018['pace_roll2Orlando']
team_2017_2018['pace_roll2Philadelphia'] + team_2017_2018['pace_roll2Washington'] + team_2017_2018['pace_roll2Milwaukee'] + team_2017_2018['pace_roll2New Orleans'] + team_2017_2018['pace_roll2Memphis'] + team_2017_2018['pace_roll2Atlanta']
team_2017_2018['pace_roll2Dallas'] + team_2017_2018['pace_roll2Denver'] + team_2017_2018['pace_roll2Utah'] + team_2017_2018['pace_roll2Minnesota'] + team_2017_2018['pace_roll2San Antonio'] + team_2017_2018['pace_roll2Portland']
team_2017_2018['pace_roll2Phoenix'] + team_2017_2018['pace_roll2Sacramento'] + team_2017_2018['pace_roll2Chicago'] + team_2017_2018['pace_roll2Toronto'] + team_2017_2018['pace_roll2New York'] + team_2017_2018['pace_roll2Oklahoma City']
team_2017_2018['pace_roll2LA Clippers'] + team_2017_2018['pace_roll2LA Lakers']




def join_pace(team,df):
    pace = pd.DataFrame(df.loc[lambda df: df.TEAMS == team])
    pace = pace[['PACE']]
    pace['pace_roll2'] = pace['PACE'].rolling(2, min_periods=1).mean().shift()
    pace['pace_roll3'] = pace['PACE'].rolling(3, min_periods=1).mean().shift()
    pace['pace_roll5'] = pace['PACE'].rolling(5, min_periods=1).mean().shift()
    pace['pace_roll10'] = pace['PACE'].rolling(10, min_periods=1).mean().shift()
    pace.drop(['PACE'], axis=1,inplace=True)
    df = df.join(pace, how='outer',rsuffix=team)
    return df



team_2017_2018 = join_pace('Boston',team_2017_2018)

team_2017_2018 = join_pace('Cleveland',team_2017_2018)



for col in team_2017_2018.columns:
    print(col)

#create new DataFrame for all of Boston's games
boston = pd.DataFrame(team_2017_2018.loc[lambda df: team_2017_2018.TEAMS == 'Boston'])
# #create new index
# boston = boston.reset_index(drop=True)
# #shift index up by 1 - so the index starts at 1
# boston.index += 1
# #Create Game# column to indicate the game of the season
# boston['GAME#'] = boston.index
# #reset index
# boston.reset_index(drop=True,inplace=True)


boston.columns

boston = boston[['PACE']]
boston['pace_roll2'] = boston['PACE'].rolling(2, min_periods=1).mean().shift()
boston['pace_roll3'] = boston['PACE'].rolling(3, min_periods=1).mean().shift()
boston['pace_roll5'] = boston['PACE'].rolling(5, min_periods=1).mean().shift()
boston['pace_roll10'] = boston['PACE'].rolling(10, min_periods=1).mean().shift()
boston.drop(['PACE'], axis=1,inplace=True)
team_2017_2018 = team_2017_2018.join(boston, how='outer')


team_2017_2018.columns

cleveland = pd.DataFrame(team_2017_2018.loc[lambda df: team_2017_2018.TEAMS == 'Cleveland'])
cleveland = cleveland[['PACE']]
cleveland['pace_roll2'] = cleveland['PACE'].rolling(2, min_periods=1).mean().shift()
cleveland['pace_roll3'] = cleveland['PACE'].rolling(3, min_periods=1).mean().shift()
cleveland['pace_roll5'] = cleveland['PACE'].rolling(5, min_periods=1).mean().shift()
cleveland['pace_roll10'] = cleveland['PACE'].rolling(10, min_periods=1).mean().shift()
cleveland.drop(['PACE'], axis=1,inplace=True)
team_2017_2018 = team_2017_2018.join(cleveland, how='left',rsuffix='CLE')


team_2017_2018 = team_2017_2018.fillna(0)

team_2017_2018.head(60)

team_2017_2018['pace_roll2_combo'] = team_2017_2018['pace_roll2'] + team_2017_2018['pace_roll2CLE']



#Looking at the rolling average of a player's minutes played at lags of 2,3,5 and 10 games

pace = boston[['PACE']]
pace['Roll2'] = pace['PACE'].rolling(2, min_periods=2).mean().shift()
pace['Roll3'] = pace['PACE'].rolling(3, min_periods=2).mean().shift()
pace['Roll5'] = pace['PACE'].rolling(5, min_periods=2).mean().shift()
pace['Roll10'] = pace['PACE'].rolling(10, min_periods=2).mean().shift()
pace['Diff2'] = abs(pace['PACE'] - pace['Roll2'])
pace['Diff3'] = abs(pace['PACE'] - pace['Roll3'])
pace['Diff5'] = abs(pace['PACE'] - pace['Roll5'])
pace['Diff10'] = abs(pace['PACE'] - pace['Roll10'])
pace.plot()
pace.describe()
# pace

pace.tail(10)





team_2017_2018.join(boston, how='outer')
```
