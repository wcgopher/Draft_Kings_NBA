```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_pickle('Prediction_first.pkl')
```

```python
data['VALUE'] = data['DraftKings_PTS'] / data['DK_SAL'] *100
data['VALUE_pred'] = data['DK_PTS_pred'] / data['DK_SAL'] *100
data['CAP_HIT'] = data['DK_SAL'] / 50000
data['CAP_VALUE'] = data['DraftKings_PTS'] / data['CAP_HIT'] / 1000
data['CAP_VALUE_pred'] = data['DK_PTS_pred'] / data['CAP_HIT'] / 1000
data['magic_number'] = data['DraftKings_PTS'] * data['CAP_VALUE']
data['magic_number_pred'] = data['DK_PTS_pred'] * data['CAP_VALUE_pred']
data['add'] = data['CAP_HIT'] * data['magic_number_pred']
data['something'] = data['VALUE'] / data['magic_number']
```

```python
data = data.drop(['DK_PTS_PER_MIN',
       'MINUTES', 'PPM_pred', 'MINS_pred'],axis=1)
```

```python
day = data.loc[(data.DATE == '02/02/2018')]
#                & (data.POSITION == 'C')]   
day.columns
```

```python
day.sort_values('something', ascending = True, inplace=True)
day
```

```python
budget = 50000
spots = 8
roster = []
positions = []
needs = {'PG': 1,'SG': 1,'SF': 1,'PF': 1,'C': 1,'PG_SG': 1,'SF_PF': 1 ,'Util': 1}
pred_output = []
output = []
spent = 0
while spots >= 1:
    for line in day.itertuples():
        if line[5] <= (budget):   #6250
            if line[2] not in roster:
                if line[4] == 'PG':
                    if needs['PG'] == 1:
                        roster.append(line[2])
                        positions.append(line[4])
                        needs['PG'] -= 1
                        break
                    if needs['PG_SG'] == 1:
                        roster.append(line[2])
                        positions.append(line[4])
                        needs['PG_SG'] -= 1
                        break
                    if needs['Util'] == 1:
                        roster.append(line[2])
                        positions.append(line[4])
                        needs['Util'] -= 1
                        break
                    else:
                        continue
                if line[4] == 'SG':
                    if needs['SG'] == 1:
                        roster.append(line[2])
                        positions.append(line[4])
                        needs['SG'] -= 1
                        break
                    if needs['PG_SG'] == 1:
                        roster.append(line[2])
                        positions.append(line[4])
                        needs['PG_SG'] -= 1
                        break
                    if needs['Util'] == 1:
                        roster.append(line[2])
                        positions.append(line[4])
                        needs['Util'] -= 1
                        break
                    else:
                        continue
                if line[4] == 'SF':
                    if needs['SF'] == 1:
                        roster.append(line[2])
                        positions.append(line[4])
                        needs['SF'] -= 1
                        break
                    if needs['SF_PF'] == 1:
                        roster.append(line[2])
                        positions.append(line[4])
                        needs['SF_PF'] -= 1
                        break
                    if needs['Util'] == 1:
                        roster.append(line[2])
                        positions.append(line[4])
                        needs['Util'] -= 1
                        break
                    else:
                        continue
                if line[4] == 'PF':
                    if needs['PF'] == 1:
                        roster.append(line[2])
                        positions.append(line[4])
                        needs['PF'] -= 1
                        break
                    if needs['SF_PF'] == 1:
                        roster.append(line[2])
                        positions.append(line[4])
                        needs['SF_PF'] -= 1
                        break
                    if needs['Util'] == 1:
                        roster.append(line[2])
                        positions.append(line[4])
                        needs['Util'] -= 1
                        break
                    else:
                        continue
                if line[4] == 'C':
                    if needs['C'] == 1:
                        roster.append(line[2])
                        positions.append(line[4])
                        needs['C'] -= 1
                        break
                    if needs['Util'] == 1:
                        roster.append(line[2])
                        positions.append(line[4])
                        needs['Util'] -= 1
                        break
                    else:
                        continue

    budget -= line[5]
    spots -= 1
    spent += line[5]
    pred_output.append(line[7])
    output.append(line[6])
    
print('Budget used: ',spent)
print(positions)
print('Predicted team output: ',sum(pred_output))
print('Actual team output: ',sum(output))
roster
```
