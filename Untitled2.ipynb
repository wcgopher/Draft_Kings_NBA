```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_pickle('player_2016_2018.pkl')
```

```python
data['DK_SAL'] = pd.to_numeric(data['DK_SAL'])
data['VALUE'] = pd.to_numeric(data['VALUE'])
data['FAVORITE'] = data['FAVORITE'].astype(int)
data['SPREAD_10+'] = data['SPREAD_10+'].astype(int)
```

```python
data.columns
```

```python
data = data.drop(['DK_SAL','DATE', 'PLAYER', 'POSITION', 'TEAM', 'OPPONENT','C_DK_PTS', 'PF_DK_PTS', 'PG_DK_PTS',
       'SF_DK_PTS', 'SG_DK_PTS', 'opp_C_allowed', 'opp_PF_allowed',
       'opp_PG_allowed', 'opp_SF_allowed', 'opp_SG_allowed','DK_PTS_PER_MIN',
                  'PACE', 'OEFF', 'DEFF','MINUTES','USAGE', 'VALUE'],axis=1)
```

```python
data = pd.get_dummies(data)
```

```python
data = data.dropna()
```

```python
data.info()
```

```python
import sklearn
from sklearn.ensemble import RandomForestRegressor

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
%matplotlib inline
```

```python
X_train, X_test = train_test_split(data, test_size=0.2)
```

```python
y_train = X_train.pop('DraftKings_PTS')
y_test = X_test.pop('DraftKings_PTS')
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
model = RandomForestRegressor(n_estimators = 100, oob_score = True, random_state = 42)

model.fit(X_train_scaled,y_train)
```

```python
model.oob_score_
```

```python
y_oob = model.oob_prediction_
```

```python
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(y_train,y_oob))
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_train,y_oob))
```

```python
print(np.sqrt(mean_squared_error(y_train,y_oob)))

```
